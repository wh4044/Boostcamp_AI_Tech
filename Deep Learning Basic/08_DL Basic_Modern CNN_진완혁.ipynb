## AlexNet
- GPU를 최대한 활용하여 분산시켜서 따로따로 트레이닝 시켰다.
- 5개의 컨볼루션 레이어, 3개의 덴스 레이어가 있다.
- Key ideas
    - ReLU activation 사용
    - GPU implementation (2 GPU) 2개의 gpu를 사용했다.
    - Local reponse normalization, Overlapping pooling
    - Data augmentation
    - Dropout


## VGGNet
- Increasing depth with 3 x 3 convolution filters(with stride 1) : 3x3 컨볼루션을 활용했다.
- Why 3 x 3 convolution?
    - 크기를 줄이는 것은 똑같지만
    - 파라미터 개수는 5 x 5에 비해 적다
    - 그래서 웬만하면 7 x 7 이하로 사용한다.


## GoogLeNet
- 1 x 1 convolution을 활용했다.

- 비슷한 구조가 네트워크 안에 반복되고 있다. 이를 network-in-network(NiN)이라고 부른다.

- 1 x 1 convolution은 channel-wise dimension을 감소시킨다.

- 1 x 1 컨볼루션을 이용하면 파라미터 개수가 약 30%가 된다.

## ResNet
- Deeper neural networks are hard to train.
    - Overfitting is usually caused by an excessive number of parameters.
    - But, not in this case.
    
    
- Add an identity map (skip connection)


- ResNet 구조를 이용하면 더 깊게 층을 쌓아도 성능이 높아진다.


## DenseNet
- DenseNet uses **concatenation** instead of **addition.**


- Dense Block
    - Each layer concatenates the feature maps of all preceding layers.
    - The number of channels increases geometrically.
    
- Transition Block
    - BatchNorm -> 1 x 1 Conv -> 2 x 2 AvgPooling
    - Dimension reduction
    
- feature map을 기하급수적으로 늘린 다음에 배치정규화, 1x1 conv, 풀링으로 차원을 줄인다.


# Summary
- Key takeaways
    - **VGG** : repeated 3 x 3 blocks
    - **GooLeNet** : 1 x 1 convolution
    - **RestNet** : skip0connection
    - **DenseNet** : concatenation
