# Sequential Model
- What makes sequential modeling a hard problem to handle?
- sequential 데이터를 다루기 어려운 이유는 데이터가 중간중간 바뀌나 빠지는 등 길이가 바뀌기 때문이다.


- 그래서 Transformer라는 구조를 사용하게 된다.
# Transformer
- **Transformer** is the first sequence transduction model based entirely on attention.
- RNN은 하나의 입력과 다른 입력이 들어가면서 은닉층에 출력이 다시 전달되는 식으로 동작했다.
- 이런 반복적인 구조를 없애고 attention이라 불리는 구조를 활용한 것이 트랜스포머이다.



- From a bird's-eye view, this is what the **Transformer** does for machine translation tasks.
- 이 방법론은 sequential 데이터를 처리하고 이 데이터를 인코딩하는 방법이다.
- 이미지 분류나 detection 등 이미지 분류에도 활용되고 있다.

- If we glide down a little bit, this is what the **Transformer** does.
- sequence를 다른 sequence로 바꾸는 것은 seq2seq 모델이라고 한다.

- 입력은 세 개의 단어로 되어있고
- 출력은 네 개의 단어로 되어있다.
- 여기서 입력과 출력의 시퀀스의 숫자가 다를 수 있다.
- 입력과 출력의 도메인이 다를 수 있다.
- 하나의 모델이 있다.


- 1. n개의 단어가 어떻게 인코더에서 처리가 되는지
- 2. 디코더와 인코더 사이에 어떤 정보를 주고받는지
- 3. 디코더가 어떻게 제너레이션 할 수 있는지 이해해야 한다.


- The **Self-Attention** in both encoder and decoder is the cornerstone of Transformer.
- 이 트랜스포머가 왜 잘 되게 되는지 나타낸다.

------------------------------------------------------

- First, we represent each word with some embedding vectors.
- 예시로 3개의 단어가 들어왔다고 하자.



- Then, Transformer encodes each word to feature vectors with **Self-Attention**
- 3개의 단어가 주어지면 3개의 벡터를 찾아준다.
- x벡터에서 z벡터로 갈 때 나머지의 x벡터들을 고려한다.


- Suppose we are encoding two words:
    - 2개의 단어만 주어졌다고 가정해보자.
    - **Thinking** and **Machines**
    
- **Self-Attention** at a high level
    - The animal didn't cross the street because ***it*** was too tired.
    - 이 문장을 이해하려면 It이 무엇에 의존하는지 알아야한다.(다른 단어와의 상호작용)
    - It이 animal과 높은 관계가 있다고 학습을 하기 때문에 더 잘 표현을하고 이해한다.


- Self-Attention 구조는 3가지 벡터를 만들어낸다.(3개의 뉴럴 네트워크)
    - Queries
    - Keys
    - Values


- Thinking 이라는 것을 찾을 때 각각의 단어마다 쿼리, 키, 밸류 벡터를 만들고
- Score 벡터라는 것도 만든다. 이는 Thinking에 대한 스코어 벡터를 계산할 때 인코딩 하고자 하는 쿼리벡터와
- 나머지 n개의 단어에 대한 키벡터를 구하여 내적을 한다.
- i번 째 단어가 나머지 n개의 단어와 얼마나 관계가 있는지(유사도가 있는지) 스코어를 구한다.


- i 번째 단어와 나머지 단어들 사이에 얼마나 상호작용을 해야하는지 알아서 학습을 시킨다.


- score 벡터를 normalization 해준다.(8로 나눠준다. => $\sqrt{key vector의 dimension}$ )
- 그러고 나서 softmax 함수를 취해준다.


- 이를 value 벡터와 weighted sum을 한 것이 Thinking 단어의 인코딩 벡터가 된다.


- 위 과정을 행렬로 표현하면 매우 간단하기 때문에 PyTorch에서 코드로 구현할 때도 간단하게 할 수 있다.
