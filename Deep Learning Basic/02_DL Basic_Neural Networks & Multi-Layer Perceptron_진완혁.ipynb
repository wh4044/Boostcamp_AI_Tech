# Neural Networks
## 'Neural networks are computing systems vaguely inspired by the biological neural networks that constitute animal brains.'




- 우리가 하늘을 날고 싶다고 해서 꼭 새처럼 날 필요는 없다. 빨리 달리고 싶다고 해서 꼭 치타처럼 달릴 필요는 없다.
- 인간의 지능을 만들고 싶다고 해서 꼭 인간의 뇌를 모방할 필요는 없다는 의미이다.
- 시작은 인간의 뇌를 모방했지만 지금의 딥러닝과 NN은 인간의 뇌를 모방해서 잘 된다고 말하기에는 너무 달라졌다.
- 굳이 딥러닝을 해석할 때 인간의 뇌를 모방했다기 보다는 이게 왜 잘 되는지 수학적으로 분석하는 것이 더 옳을 수 있다.

## Neural networks are function approximators that stack affine transformations followed by nonlinear transformations.
- affine transformations에다가 비선형 연산이 반복적으로 일어나는 모델을 뉴럴 네트워크라고 부르겠다.


# Linear Neural Networks
- 가장 간단한 뉴럴 네트워크. 입력과 출력이 각각 1차원.
- 선에 대한 기울기와 절편 파라미터를 찾는 것


**Loss** : 우리가 이루고자 하는 현상이 이루어졌을 때 줄어드는 함수

- We compute the partial derivatives w.r.t the optimization variables.
- 우리가 사용할 전략 : 역전파(backpropagation)
- 파라미터가 어느 방향으로 움직였을 때 loss가 줄어드는지 찾고 그 방향으로 파라미터를 바꾼다.
- loss function을 w로 편미분하여 나온 값에 적절한 값을 곱하여 기존의 w에서 빼준다.
- loss function을 b로 편미분하여 나온 값에 적절한 값을 곱하여 기존의 b에서 빼준다.



- Of course, we can handle multi dimensional input and output.
- W와 b가 행렬과 벡터로 돼있다.
