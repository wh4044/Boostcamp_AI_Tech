# Careers in AI
AI를 직업으로 가지려면 어디를 가야할까?

대부분 학교와 회사 중에 고민을 한다.

- 두 조직의 목표의 차이를 이해하자.
    - 학교 => 논문을 써서 연구 성과 만들기
    - 회사 => 서비스/상품을 만들어서 돈을 버는 곳


- '논문을 쓰자' : 학교로~!
    - 회사는 논물 쓸 시간이 없다.
    - 학교는 논문을 쓰는 방식에 지도를 받을 수 있다.
    
    
- '상품/서비스를 만들자' : 회사로~!
    - 회사는 상대적으로 데이터와 계산자원이 풍부하다.
    - 어마어마한 양의 트래픽을 통해 발생되는 대규모의 리얼 데이터를 만질 수 있다.
    
## AI를 다루는 회사의 종류
**1. AI for X**
- AI로 기존 비즈니스를 더 잘하려는 회사
- 비용을 줄이거나, 품질을 높이기 위함
- AI는 보조 수단. 대부분의 회사가 해당한다.
    

**2. AI centric**
- AI로 새로운 비즈니스를 창출    
- 새로운 가치창출을 위함    
- 신생 회사가 많다.
    
## AI를 다루는 팀의 구성
**1. Business**
- 사업 기획자, 서비스 기획자, 마케팅/세일즈/PR, etc...

**2. Engineering**
- Data engineer, AI engineer, AlOps engineer, etc...

AI 팀에서 엔지니어가 되면 보통 논문을 읽고 모델을 학습하는 일을 떠올린다.

하지만 현실적으로는 훨씬 더 다양한 일을 한다.

AI/ML 모델링은 팀 전체 업무의 **일부**. 다양한 업무가 있는 만큼 팀 내부에는 다양한 역할이 있다.


현실에서는 정말 다양한 역할이 있고 **100% 하나의 포지션의 역할을 하는 경우는 굉장히 드물다**

아래 그림에서 가로축은 '업무의 범위', 세로축은 '업무의 깊이'








**회사들은 보통 비어있거나 보강하고 싶은 곳이 생기면 공고를 내거나 추천을 받아 사람을 뽑는다**

## AI 엔지니어링 포지션의 트렌드
- AI modeling 외에도 다양한 포지션이 많다.
- 오히려 AI modeling에 대한 수요는 점차 줄어들 가능성도 있다.

# How to start my AI engineering career
**먼저 자기 스스로에 대한 이해가 필요하다.**
- ex)
    - 나는 비즈니에스 관심이 있는가?
    - 내가 만든 모델을 사람들이 쓰면서 실생활에 변화가 있기를 바라는가?
    - 나는 AI/ML 모델링 뿐만 아니라 그 과정에서 발생하는 모든 일(웹프로그래밍, 데이터 전처리)을 기꺼이 할 수 있는가?
    
    - 엔지니어로써의 커리어를 쌓아볼 수도 있다.

**확장을 할 때는 일찍 시작하는 것을 권장**
- 학교 다니면서 AI 관련 인턴십/알바를 적극 권장
    - 각 포지션의 차이를 이해할 수 있다.
    - 각 팀이 무엇에 관심을 가지는지 체감.
    - '일'하는 방법을 알게 된다.

## 인턴쉽 대신 할 수 있는 것.
- Ai competition(ex. 캐글)
    - Forum이 활발하고 내용이 풍부하다!
    - 머신러닝은 hands-on으로 문제를 풀어보는 게 중요함.
    
- 최신 논문 재현
    - 장점 : 논문에는 빠져있는 내용이 많다. 그렇기 때문에 구현 만으로도 많은 백그라운드 지식을 학습할 기회가 생김.
    - 유의점 : 업무에서 이 능력만 활용하는 사람은 드물다.

## 회사를 가기위해 석박사 학위를 받아야 하는가?
- **필수**는 아니다.
- 다양한 곳에 지원하고, 내가 가장 많이 성장할 수 있는 환경을 택하자.
    - 배울 수 있는 사람이 있는 곳
    - 풍부한 데이터/계산 자원이 있는 곳
    - 집중할 수 있는 문화 등
    - 나만의 기준을 고려해서 선택을 하자!
    
**꼭 컴공과를 나오지 않아도 된다.**
- 참고 : Upstage AI research engineer 12명의 학부 출신 학과
    - 컴퓨터공학, 산업공학, 화학생명공학, 영어영문, 심리학 등등 다양하다!

## ~~을 꼭 배워야 하는가?
- 한 가지만 깊게 막 파기보다는, 여러가지 범위를 확장해 가면서 필요할 때 그 분야를 집중적으로 파도 된다!

## 어떤 역량을 쌓아야할까?
- 내가 회사의 사장이 되었다고 생각해보자.
- 어떤 사람과 일을 하고 싶은가??
    - AI를 떠나 '일'을 잘하는 사람과 일하고 싶다.
        - 말, 듣, 쓰를 잘 하는 사람.
        - 일처리가 빠르고 꼼꼼하고 정확한 사람.
        
**공통적인 기술적/학문적 역량**
- 컴퓨터 공학에 대한 기본적인 이해와 소프트웨어 엔지니어링 능력
    - AI/ML에 대한 이해가 아주 특출난 경우가 아니라면 소프트웨어 엔지니어링 실력이 바탕이 되는 것이 중요하다.
    
    - 최신 기술을 빠르게 습득하기 위해 영어 독해 능력이 중요하다.(논문)

**공통적인 Soft_skills**
1. 끈기
2. 겸손
3. 열정
4. 협동
5. 친절

## 나의 역량을 보여주는 방법
**짧은 길이의 이력서 안에 상대방에게 좋은 인상을 남길 수 있어야 한다.**
- 팀으로 일하기 때뭉네 모든 것을 다 잘할 필요가 없다.
- 강력한 한 방이 중요!!


- 실력을 보여줄 수 있는 '스펙'의 예시
    - Coding competitions: ICPC 등 규모가 큰 프로그래밍 대회에서 입상 경력이 있는지
    - AI competitions: Kaggle 등 규모가 큰 AI 관련 대회에서 수상한 경력이 있는지
    - Publication record: NeurIPS 등 AI 관련 주요 국제 학회나 워크샵에 1저자로 출판한 논문이 있는지
    - 서비스 경험: 경력자의 경우 실제로 라지 스케일 서비스를 다뤄봤으며 그 과정을 주도했는지
    - 다른 회사 경력: 다른 AI 관련 회사의 근무 경험과 성과


# Summary
- AI 관련 포지션의 이름은 완전히 합의되지 않았기 때문에 회사마다 다른 이름이더라도 같은 일을 하는 경우가 많다.


- 각 사람이 100% 하나의 역할을 수행하는 경우는 굉장히 드물다. 팀으로 활동(내가 팀의 빈 공간을 채우자!)


- 나의 관심사와 능력 뿐 아니라 시장의 흐름과 수요/공급을 고려하여 커리어 방향 정하기!


- 모든 것을 잘하려고 하지말고 팀에 기여할 수 있는 능력을 키우자!
