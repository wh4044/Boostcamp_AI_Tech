# Text 사용하기
## Text in Viz
- Visual representation들이 **줄 수 없는 많은 설명을 추가**해줄 수 있다.
- 잘못된 전달에서 생기는 **오해를 방지**할 수 있다.


- 하지만 Text를 과하게 사용하면 오히려 이해하는데에 방해가 된다.


- Matplotlib에서 제공하는 API를 바탕으로 Text를 이해하자!

## Anotomy of a Figure(Text Ver.)

- **Title** : 가장 큰 주제를 설명한다.


- **Label** : 축에 해당하는 데이터 정보를 제공한다.


- **Tick Label** : 축에 눈금을 사용하여 스케일 정보를 추가한다.


- **Legend** : 한 그래프에서 2개 이상의 서로 다른 데이터를 분류하기 위해서 사용하는 보조 정보이다.


- **Annotation(Text)** : 그 외의 시각화에 대한 설명을 추가한다.
