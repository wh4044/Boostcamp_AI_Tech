# Pie Chart

## Pie Chart
- 원을 부채꼴로 분할하여 표현하는 통계 차트이다.
    - 전체를 백분위로 나타낼 때 유용하다.
    

- 가장 많이 사용하는 차트지만 지양한다.
    - 비교하기 어렵다.
    - 유용성이 떨어진다.
    - 오히려 bar pot이 더 유용하다(각도 <<< 길이)
        - 사용을 꼭 하고 싶다면 함께 사용할 것을 권장한다.
        - 후에 waffle chart가 더 비교에 유용하다.


# Pie Chart 응용
- Pie Chart와 유사한 Chart들

## Donut Chart
- 중간이 비어있는 Pie Chart
    - 디자인적으로 선호되는 편이다.
    - 인포그래픽에서 종종 사용한다.
    - Plotly에서 쉽게 사용 가능하다.


## Sunburst Chart
- 햇살(sunburst)을 닮은 차트

- 계층적 데이터를 시각화하는 데 사용
    - 구현 난이도에 비해 화려하다는 장점이 있다.
    - 오히려 Treemap을 추천한다.
    - Plotly로 쉽게 사용 가능하다.
