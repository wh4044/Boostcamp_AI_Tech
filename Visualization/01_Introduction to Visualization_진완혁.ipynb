# '데이터' 시각화

- 데이터 시각화를 위해서는 **데이터**가 우선적으로 필요하다.

- 데이터셋의 종류
> 1. 정형데이터
> 2. 시계열 데이터
> 3. 지리데이터
> 4. 관계형(네트워크)데이터
> 5. 계층적 데이터
> 6. 다양한 비정형 데이터

## 정형데이터

- 테이블 형태로 제공되는 데이터, 일반적으로 csv, tsv 파일로 제공된다.
- Row가 데이터 1개 **item**
- Column은 **attribute(feature)**

- 가장 쉽게 생각할 수 있는 데이터셋
> - 통계적 틍성과 feature사이 관계
> - 데이터 간 관계
> - 데이터 간 비교

## 시계열데이터

- 시간 흐름에 따른 데이터를 Time_Series
- 기온, 주가 등 정형데이터와 음성, 비디오와 같은 비정형 데이터 존재
- 시간 흐름에 따른 **추세(Trend), 계절성(Seasonality), 주기성(Cycle)** 등을 살핌

## 지리/지도 데이터

위도, 경도에 대한 정보도 알아야한다.
- 지도 정보와 보고자 하는 정보 간의 조화 중요 + 지도 정보를 단순화 시키는 경우도 존재
- 거리, 경로, 분포 등 다양한 실사용

## 관계 데이터

json타입이나 linked list로 전달될 수 있다. 가독성이 좋지 않아서 엄청 많은 노드에 있어 어떻게 객체를 잘 표시하고 관계를 표시할지 매핑방법이 중요하다!
- 객체와 객체 간의 관계를 시각화
> - Graph Visualization / Network Visualization

- 객체는 Node로, 관계는 Link로 구성
- 크기, 색, 수 등으로 객체와 가중치를 표현한다.
- 휴리스틱하게 노드 배치를 구성한다.

## 계층적 데이터

회사 조직도, 가계도 등 계층이 있는 데이터를 말한다. 계층을 강조할 때 사용한다.
- 관계 중에서도 포함관계가 분명한 데이터이다.
> - 네트워크 시각화로도 표현 가능
- Tree, Treemap, Sunburst 등이 대표적이다.

----

데이터는 대표적으로 4가지로 분류된다.
- 수치형(numerical)
>- 연속형(continuous) : 길이, 무게, 온도 등
>- 이산형(discrete) : 주사위 눈금, 사람 수 등

- 범주형(categorical)
>- 명목형(nominal) : 혈액형, 종교 등(순서가 중요하지 않음)
>- 순서형(ordinal) : 학년, 별점, 등급 등

데이터의 종류에 따라 어떤 그래프를 이용하여 시각화를 할지 감을 잡아야한다!

# 시각화 이해하기
## 마크와 채널

- A **mark** is a basic graphical element in an image
> 점, 선, 면으로 이루어진 데이터 시각화

- A visual **channel** is a way to control the appearance of marks, independent of the dimensionality of the geometric primitive.
> 각 마크를 변경할 수 있는 요소들

## 전주의적 속성

- 기울기, 길이, 너비, 크기 등이 다르면 한 눈에 알아챌 수 있는 속성을 전주의적 속성이라고 한다.
- 그러나 동시에 사용하면 인지하기가 어렵다.
- 시각적 분리(visual pop-out)가 잘 일어나는 시각화가 가독성이 좋고 전하고자 하는 인사이트를 잘 전달할 수 있다.
