# Grid 이해하기
## Default Grid
- 기본적인 **Grid**는 축과 평행한 선을 사용하여 거리 및 값 정보를 보조적으로 제공한다.


- 색은 다른 표현들을 방해하지 않도록 **무채색**을 사용한다. (`color`)


- 항상 Layer 순서 상 **맨 밑**에 오도록 조정한다. (`zorder`)


- **큰** 격자/**세부** 격자 (`which='major'`, `'minor'`, `'both'`)


- X축? Y축? 동시에 (`axis='x'`, `'both'`)

## 다양한 타입의 Grid
- 전형적인 Grid는 아니지만 여러 형태의 Grid가 존재한다.
    - 두 변수의 합이 중요하다면 **x+y = c**
    - 비율이 중요하다면 **y = cs**
    - 두 변수의 곱이 중요하다면 **xy = c**
    - 특정 데이터를 중심으로 보고 싶다면 **(x-x')^2 + (y-y')^2 = c**

### X + Y = C를 사용한 Grid

- 회색 선에 걸치는 값은 X+Y 값이 동일하다.


- Feature의 절대적 합이 주용한 경우
    - 공격 + 수비 합으로 평가
    - 국어 + 수학 비중 평가


### Y = CS를 사용한 Grid

- 가파를 수록 Y/x가 커진다.


- Feature의 비율이 중요한 경우.

### 동심원을 사용

- 특정 지점에서 거리를 살펴볼 수 있다.


- 가장 가까운 포인트를 찾거나


- 한 데이터에서 특정 범위의 데이터를 확인할 수 있다.
