# Facet
## Mulitple View
- **Facet**이란 분할을 의미한다


- 화면 상에 View를 분할 및 추가하여 다양한 관점을 전달한다.
    - 같은 데이터셋에 **서로 다른 인코딩**을 통해 다른 인사이트를 보여준다.
    - **같은 방법으로 동시에** 여러 feature를 보거나
    - 큰 틀에서 볼 수 없는 **부분 집합을 세세**하게 보여줄 수 있다.

# Matplotlib에서 구현
## Figure와 Axes

- Figure에는 큰 틀, Ax는 각 플롯이 들어가는 공간이다.


- Figure는 언제가 1개, 플롯은 N개이다.

## N x M subplots
- 가장 쉬운 방법은 3가지 방법이다.
    - `plt.subplot()`
    - `plt.figure` + `fig.add_subplot()`
    - `plt.subplots()`
    
    
- 쉽게 조정할 수 있는 요소
    - `figuresize`
    - `dpi`
    - `sharex`, `sharey`
    - `squeeze`
    - `aspect`

## Grid Spec의 활용

- 그리드 형태의 subplots


- 기존 Subplots로 4x4를 만들 수 있다.


- 여기서 다른 사이즈를 만들기 위해서는!?


- 이 서브플롯을 표현하기 위해서 2가지 방법이 있다.
    1. Slicing 사용
    2. x, y, dx, dy를 사용
    
    
- 1. Numpy의 slicing을 생각해보면

    파랑 : axes[0, :3]
    
    노랑 : axes[1:, :3]
    
    보라 : axes[3, :]
    
   
`fig.add_grid_spec()`

- 2. (x, y), dx, dy 포맷으로 작성하면

    파랑 : (0, 0), 1, 3
    
    노랑 : (1, 0), 3, 3
    
    보라 : (0, 3), 1, 4
    
`fig.subplot2grid()`

## 내부에 그리기

- Ax 내부에서 서브플롯을 추가하는 방법이다.
    - 미니맵과 같은 형태로 추가하거나
    - 외부 정보를 적은 비중으로 추가한다.

`ax.inset_axes()`


- 그리드를 사용하지 않고 사이드에 추가한다.
    - 단, 방향의 통계정보를 제공할 수도 있다.
    - 제목 등의 텍스트 추가도 가능하다.
    
`make_axes_locatable(ax)`
