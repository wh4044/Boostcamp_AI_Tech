# Interactive를 사용하는 이유
## 정적 시각화의 단점
- 정형 데이터에서 Feature가 10개 있다면
    - 각각을 살펴보는 데 10개의 plot이 필요하다.
    - 각각의 관계를 살펴보는데 10*9 / 2 = 45개의 plot 필요
    - 가볍게 살펴보는 것도 약 50개의 plot이 필요하다.
    - **공간적 낭비**가 크다!!
    
    
- 각각의 사용자는 원하는 인사이트가 다를 수 있다.
    - 필요한 인터랙션을 통해 원하는 정보를 얻을 수도 있다.
    - 물론 설득을 위해서 **원하는 메세지를 압축해서 담는 것은 정적 시각화**의 장점이다.
    

## 라이브러리 소개
- 대표 라이브러리 3개
    - **Plotly**
    - **Bokeh***
    - **Altair**
    
    
- 문법과 제공하는 방법들이 차이가 있다.
    - 원하는 라이브러리를 선택하면 된다!

# Interactive Viz Libraray
## Matplotlib
- Matplotlib도 인터랙티브를 제공한다.


- 단, 주피터 노트북 환경 또는 Local에서만 실행할 수 있다.
    - 다른 라이브러리들은 웹에 deploy 가능하다.
    - **mpld3**를 사용하면 웹에 D3-based Viewer를 제공한다.
    
## Plotly
- 인터랙티브 시각화에 가장 많이 사용되는 Plotly
    - Python 뿐만이 아니라 R, JS에서도 제공하고 있다.
    
- 예시 + 문서화가 잘 되어있다.
- 통계 시각화 외에도 지리 시각화 + 3D 시각화 + 금융 시각화 등 다양한 시각화 기능을 제공한다.
- JS 시각화 라이브러리 D3js를 기반으로 만들어져 웹에서 사용가능하다.
- 형광 Color가 인상적이다.

## Plotly Express
- Plotly를 seaborn과 유사하게 만들어 쉬운 문법이다.
- 커스텀 부분이 부족하지만 다양한 함수를 제공한다.


## Bokeh
- 문법은 Matplotlib과 유사한 부분이 있다.


- 기본 Theme이 Plotly에 비해 깔끔하다.
- *비교적으로 문서화가 부족하다.*


## Altair
- Vega 라이브러리를 사용하여 만든 인터랙티브이다.
- 시각화를 + 연산 등으로 배치하는 것이 특징이다.


- 문법이 Pythonic하지 않다.(js스러움.)
- 데이터 크기에 5000개 제한이 있다.


- Bar, Line, Scatter, Histogram에 특화되어 있다.
