# Seaborn이란?
## seaborn 소개
- **seaborn**은 Matplotlib 기반 통계 시각화 라이브러리이다.
    - 통계 정보 : 구성, 분포, 관계 등
    - Matplotlib 기반이라 **Matplotlib으로 커스텀** 가능하다.
    - **쉬운 문법**과 **깔끔한 디자인**이 특징이다.
    


- 디테일한 커스텀보다는
    - **새로운 방법론**을 위주로
    - 앞서 공부한 **이론과의 연결**을 중심으로 한다.

## Seaborn 설치
- 아직 버전 1.0이 나오지 않아 언제든 업데이트 가능성이 존재한다.


- 강의에서는 버전 **0.11**로 진행한다.
    - Kaggle, Google Colab에서 버전이 다를 수 있으니 꼭 설치해야 한다.
    
    
- `pip install seaborn==0.11`
- `import seaborn as sns`
    - 왜 sns인지 아는 사람은 없다...!

## 다양한 API
- Seaborn은 시각화의 목적과 방법에 따라 API를 분류하여 제공하고 있다.
    - Categorical API
    - Distribution API
    - Relational API
    - Regression API
    - Multiples API
    - Theme API
    
    
- 모든 내용을 소개하기 보다는 필요한 내용을 부분적으로 설명함!
