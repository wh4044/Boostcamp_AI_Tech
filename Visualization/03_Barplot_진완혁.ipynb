# Bar plot이란?
- **Bar plot**이란 직사각형 막대를 사용하여 데이터의 값을 표현하는 차트나 그래프를 말한다.


- 막대그래프, bar chart, bar graph 등의 이름으로 사용된다.


- 범주(category)에 따른 수치 값을 비교하기에 적합한 방법이다.
    - 개별 비교, 그룹 비교 모두 적합하다.
    
    
- 막대의 방향에 따라 분류한다. (.bar() / .barh())
    - 수직(vertical) : x축에 범주, y축에 값을 표기한다.(default)
    - 수평(horizontal) : yㅊ축에 범주, x축에 값을 표기한다.(범주가 많을 때 적합하다.)


## Multiple Bar Plot

- 임의의 그룹 데이터를 시각화
    - A, B, C, D, E 데이터가 두 그룹에 존재
    - Group Sky : [1, 2, 3, 4, 5]
    - Group Pink : [4, 3, 2, 5, 1]
    
**1. 플롯을 여러 개 그리는 방법**

**2. 한 개의 플롯에 동시에 나타내는 방법**

    1. 쌓아서 표현하기
    2. 겹쳐서 표현하기
    3. 이웃에 배치하여 표현하기
    


## Stacked Bar plot
- 2개 이상의 그룹을 쌓아서 표현하는 bar plot
    - 각 bar에서 나타나는 그룹의 순서는 항상 유지된다.
    
    
- 맨 밑의 bar의 분포는 파악하기 쉽지만
    - 그 외의 분포들은 파악하기 어렵다.
    - 2개의 그룹이 positive / negative라면 축 조정이 가능하다.
    
    
- .bar()에서는 bottom 파라미터를 사용


- .barh()에서는 left 파라미터를 사용

스카이와 핑크가 같은지 다른지 한 눈에 알기 힘들다.

### Percentage Stacked Bar Chart : 전체에서 비율을 나타낸다.


## Overlapped Bar Plot
- 2개 그룹만 비교한다면 겹쳐서 만드는 것도 하나의 선택지
    - 3개 이상에서는 파악이 어렵기 때문이다.
    
    
- 같은 축을 사용하니 비교가 쉽다.
    - 투명도를 조정하여 겹치는 부분 파악 (alpha)
    
    
- Bar plot보다는 Area plot에서 더 효과적이다.


## Grouped Bar plot

- 그룹별 범주에 따른 bar를 이웃되게 배치하는 방법이다.


- Matplotlib으로는 비교적 구현이 까다롭다.(수치형으로 조정을 해야하기 때문이다.)
    - 적당한 테크닉(.set_xticks(), .set_xticklabels())
    
    
- 앞서 소개한 내용 모두 그룹이 5개 ~ 7개 이하일 때 효과적이다.


- 그룹이 많다면 적은 그룹은 ETC로 처리한다. 그룹이 너무 많아도 안 좋다. 막대가 너무 많으면 알아보기도 힘들다.


# 정확한 Bar Plot

## Principle of Proportion Ink
- 실제 값과 그에 표현되는 그래픽으로 표현되는 잉크 양은 비례해야 한다.


- 반드시 x축의 시작은 **zero(0)!!**
    - 만약 차이를 나타내고 싶다면 plot의 세로 비율을 늘리기
    
    
- 막대 그래프에 한정되는 원칙은 아니다.


가독성은 왼쪽이 좋지만 사람들에게 인지의 방해를 주지 않기 위해서는 축을 x축으로 잡아주고 차이를 보여주고 싶으면 세로축의 비율을 늘려준다.

## 데이터 정렬하기
- 더 정확한 정보를 전달하기 위해서는 정렬이 필수이다.
    - Pandas에서는 sort_values(), sort_index()를 사용하여 정렬한다.
    
    
- 데이터의 종류에 따라 다음 기준으로
    1. 시계열 | 시간순
    2. 수치형 | 크기순
    3. 순서형 | 범주의 순서대로
    4. 명목형 | 범주의 값 따라 정렬
    
    
- 여러 가지 기준으로 정렬을 하여 패턴을 발견한다.


- 대시보드에서는 interactive로 제공하는 것이 유용하다.

## 적절한 공간 활용
- 여백과 공간만 조정해도 가독성이 높아진다.


- Matplotlib의 bar plot은 ax에 꽉 차서 살짝 답답하다.


- Matplotlib techniques
    - X/Y axis Limit(.set_xlim(), .set_ylim())
    - Spines(.spines[spine].set_visible())
    - Gap(width)
    - Legend(.legend())
    - Margins(.margins())

## ETC
- 오차 막대를 추가하여 Uncertainty 정보를 추가 가능하다.(errorbar)


- Bar 사이 Gap이 0이라면 -> **히스토그램(Histogram)**
    - .hist()를 사용하여 가능
    - 연속된 느낌을 줄 수 있다.
    

- 다양한 Text 정보 활용하기
    - 제목(.set_title())
    - 라벨(.set_xlabel(), .set_ylabel())
