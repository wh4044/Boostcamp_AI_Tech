# Polar Plot
- 극 좌표계를 사용한다.

## Polar Plot
- 극 좌표계(Polar Coordinate)를 사용하는 시각화이다.
    - 거리(R), 각(Theta)을 사용하여 plot을 그린다.
    

- 회전, 주기성 등을 표현하기에 적합하다.


- **projection = polar** 을 추가하여 사용한다.


- 해당 그래프는 Scatter
    - Line, Bar 모두 가능하다.


## Data Converting
- 이미 앞서 사용한 방식(Grid 등)


- 직교 좌표계 X, Y에서 변환 가능하다.
    - X = R cos$\theta$
    - Y = R sin$\theta$


# Radar Pot
- 극 좌표계에서 가장 대표적으로 사용할 수 있는 Radar Plot이다.

## Rader Chart
- 극 좌표계를 사용하는 대표적인 차트이다.


- 별 모양으로 생겨 Star Plot으로 불리기도 한다.


- 중심정을 기준으로 N개의 변수 값을 표현할 수 있다.


- 데이터의 Quality를 표현하기에 좋다.
    - 캐릭터의 강함
    - 운동 선수 분석(게임 및 방송 등)
    - 비교에도 적합하다.


## Radar Chart 주의점
- 각 feature는 독립적이며, 척도가 같아야 한다.
    - 순서형 변수와 수치형 변수가 함께 있다면 고려가 필요하다.    
    
    - EX) 자동차를 예시로 들면
        - 안정성 평가, 소비자 만족도 등 5점 만점
        - 최대 속도, 연비 등 절대적 수치
        - 과연 하나의 Radar Plot에 표현하는 게 맞을까?
        
        
- 다각형의 면적이 중요해보지만 feature의 순서에 따라 많이 달라진다.
- feature 이름을 정렬 했는지 안 했는지에 따라 다르게 보인다.


- Feature가 많아질수록 가독성이 떨어진다.
