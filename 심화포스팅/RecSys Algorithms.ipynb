추천 시스템과 추천 알고리즘은 밀접한 관계가 있다.

추천 시스템 : 아이템이 있고 바탕으로 어떤 추천을 할지, 결과물을 어떻게 보여줌으로써 어떤 도움을 줄 수 있을지를 총괄하는 것.

추천 알고리즘 : 아이템 풀에서 후보군들을 만들고 후보군들 중에서 인텐션이나 필터링을 하여 랭킹이나 예측을 하는 것.

# Contents-based Recommender System(컨텐츠기반 추천시스템)

- 사용자가 과거에 좋아했던 아이템 파악 -> 그 아이템과 비슷한 아이템을 추천
- EX) 아이언맨 영화에 높은 평점(4.7점)을 줬던 사람 -> 라라랜드보다 캡틴 마블을 더 좋아할 것이다!



# Collaborative Filtering(협업 필터링)
- 나와 비슷한 성향이나 취향을 갖는 유저가 좋아했던 아이템을 나에게 추천.
- 간단하면서 수준 높은 정확도를 나타낸다.
- EX) 아이언맨에 4.5점을 준 유저 A, B => 과거에 유저 A가 좋은 평가를 준 캡틴 마블을 유저 B에게 추천.



# Hybrid Recommender System
- 컨텐츠기반과 협업필터링은 서로 장단점이 있다.


- 그렇기 때문에 두 가지를 융합하여 장점을 활용하면 더 좋은 추천을 할 수 있다.


- Collaborative Filtering은 새로운 아이템에 대한 추천이 부족하다. 
    - => A와 B가 동일하게 시청하고 평가를 해야하기 때문에 특정 후보군들이 정해져 있을 수 밖에 없다.
    
    
- Content-based를 활용하면 cold-start 문제에 도움을 줄 수 있다.
    - cold start : 유저/아이템에 대해서 정보가 없거나 적은 경우에는 정확한 추천이 이루어 질 수가 없는 것.
    
    
즉, Collaborative Filtering만 사용하면 새로운 컨텐츠에 대한 추천은 부족하지만, Content-based에서 부족한 부분을 채워서 더 풍부한 추천을 할 수 있다.

# Context-based Recommendation
- 유저가 어떤 위치에 있는지에 따른 Location-based 추천시스템


- 시간에 따른 추천시스템


- 상황이나 맥락에 따라 적절한 추천을 해주는 것이다.

# Community-based Recommendation
- SNS 등의 네트워크 데이터를 활용한다.


- 사용자가 어떤 친구 또는 속한 커뮤니티에 따라 선호도를 추천한다.


- 인스타그램에서 어떤 유저를 팔로우하고 커뮤니티에 속해있는지에 대한 데이터를 활용하여 추천시스템 적용

# Knowledge-based Recommendation
- 특정 도메인 지식을 바탕으로 아이템의 features를 활용한 추천이다.


- Constraint-based Recommendation
    - 추천 시스템이 추천 알고리즘을 통하여 점수를 계산하고 결과 리스트가 나왔을 때, 정해진 규칙이 존재하면 그 리스트를 규칙에 적용하여 최종적으로 결과를 산출한다.
    

- Case-based Recommendation
    - 사용자의 현재 문제나 해결책 중에 가장 적합한 것을 추천해주는 것이다. 특정 상황에서 특정 사건에 맞게 추천을 할 수 있다.
